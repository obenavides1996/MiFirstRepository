```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import joblib
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import cross_validate
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt
from sklearn.inspection import permutation_importance
from sklearn.preprocessing import *
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
train = pd.read_csv("s3://germancredit2/datos/train/train.csv",index_col=0)
test = pd.read_csv("s3://germancredit2/datos/train/test.csv",index_col=0)
```

```python
sub_train , val = train_test_split(train,test_size = 0.3, random_state =42)
```

```python
X_train = sub_train.drop("Risk",axis=1)

X_val = pd.get_dummies(val.drop("Risk",axis=1))
```

```python
y_train = sub_train["Risk"]
y_val = val["Risk"]
```

```python
X_train = pd.get_dummies(X_train)
X_train_scaled = MinMaxScaler().fit_transform(X_train)
X_val_scaled= MinMaxScaler().fit_transform(X_val)
```

```python
y_train
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
result = permutation_importance(model, X_val, y_val, n_repeats=10)
```

```python
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
```

```python
from sklearn import metrics
```

```python
array_train =[]
array_val = []
mean_importance = []
for train_indices, val_indices in k_fold.split(X_train,y_train):
    model = DecisionTreeClassifier(max_depth=3)
    
    X_sub_train = X_train.iloc[train_indices,:]
    y_sub_train = y_train.iloc[train_indices]
    X_val = X_train.iloc[val_indices,:]
    y_val = y_train.iloc[val_indices]
    
    model.fit(X_sub_train,y_sub_train)
    
    y_predict_train = model.predict(X_sub_train)
    y_predict_val = model.predict(X_val)
    
    array_train.append(metrics.accuracy_score(y_predict_train,y_sub_train))
    array_val.append(metrics.accuracy_score(y_predict_val,y_val))
    result = permutation_importance(model, X_val, y_val, n_repeats=10)
    plt.figure(figsize=(20,10))
    plt.bar(x=X_sub_train.columns,height=result["importances_mean"])
    plt.xticks(rotation=90)
    plt.grid()
    plt.show()
    
```
